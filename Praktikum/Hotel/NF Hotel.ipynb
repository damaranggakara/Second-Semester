```python
#HOTEL
#Import
import pandas as pd
import numpy as np
```

```python
# Read the Excel file into a pandas DataFrame
filedata = r"C:\Users\umay2\Desktop\SUFFER\UNIV\SEMESTER 2\Basis Data\PRAKTIKUM BASDAT\Tugas 1\Hotel\hotel_excelvr_0nf.xlsx"
readexcel = pd.read_excel(filedata)
df= pd.DataFrame(readexcel)
print("DATA INVENTORY")
print(df)
```

```python
#1NF
#Cek bila ada data yang kosong
onenf= pd.isnull(df).sum()
print(onenf)
```

```python
#Isi kolum meal dengan Meal Plan 1
for i in range(len(df['meal'])):
    if pd.isnull(df.loc[i, 'meal'])==True:
        df.loc[i,'meal']="Meal Plan 1"
#Isi kolum room type dengan Room_Type 1
for i in range(len(df['room type'])):
    if pd.isnull(df.loc[i, 'room type'])==True:
        df.loc[i,'room type']="Room_Type 1"
#transfer hasil 1NF
df.to_excel("C:\\Users\\umay2\\Desktop\\SUFFER\\UNIV\\SEMESTER 2\\Basis Data\\PRAKTIKUM BASDAT\\Tugas 1\\Hotel\\hotel_excelvr_1nf.xlsx")
```

```python
#2NF
#Hilagkan partial dependency 
#membuat 2 kelompok kolum terdiri dari detail guest dan booking details
guestid = df[['Booking_ID','Guest_ID', 'adults', 'children','meal','parking','requests']]
bookingid= df[['Booking_ID','weekend nights', 'week nights','room type','lead time',
               'price','date of reservation','booking status']]
#hasil 2nf
print("GUESTS DETAILS")
print(bookingid)
print("BOOKING DETAILS")
print(guestid)
#transfer hasil 2nf
with pd.ExcelWriter("C:\\Users\\umay2\\Desktop\\SUFFER\\UNIV\\SEMESTER 2\\Basis Data\\PRAKTIKUM BASDAT\\Tugas 1\\Hotel\\hotel_excelvr_2nf.xlsx") as writer:
    bookingid.to_excel(writer, sheet_name="GUESTS DETAILS")
    guestid.to_excel(writer, sheet_name="BOOKING DETAILS")
```

```python
#3NF
#Hilagkan transitive dependency 
#membuat 3 kelompok kolum
status = df[['Guest_ID','Booking_ID', 'booking status']]
guestid = df[['Guest_ID','Booking_ID', 'adults', 'children','meal','parking','requests']]
bookingid= df[['Booking_ID','weekend nights', 'week nights','room type','lead time',
               'price','date of reservation','booking status']]
#hasil 3nf
print("STATUS")
print(status)
print("GUESTS DETAILS")
print(bookingid)
print("BOOKING DETAILS")
print(guestid)
#transfer hasil 3nf
with pd.ExcelWriter("C:\\Users\\umay2\\Desktop\\SUFFER\\UNIV\\SEMESTER 2\\Basis Data\\PRAKTIKUM BASDAT\\Tugas 1\\Hotel\\hotel_excelvr_3nf.xlsx") as writer:
    status.to_excel(writer, sheet_name="STATUS")
    bookingid.to_excel(writer, sheet_name="BOOKING DETAILS")
    guestid.to_excel(writer, sheet_name="GUEST DETAILS")
```
