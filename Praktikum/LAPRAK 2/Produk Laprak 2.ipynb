```python
#Soal 1
sql= "SELECT * FROM productline.data_laprak_2 WHERE STR_TO_DATE(date, '%m/%d/%Y') BETWEEN '2019-05-01' AND '2019-05-31';"
kursor1.execute(sql)
result=kursor1.fetchall()
for row in result:
    print(row)
```

```python
#Soal 2
sql = """
SELECT 
    Product_line, 
    SUM(Total) AS total_produk
FROM 
    productline.data_laprak_2
GROUP BY 
    Product_line
ORDER BY 
    total_produk DESC;
"""
kursor1.execute(sql)
result_total_produk = kursor1.fetchall()
print("\nTotal pemasukan setiap kategori")
print("-" * 50)
for row in result_total_produk:
    print(f"Kategori: {row[0]}")
    print(f"Total kategori: {row[1]:.2f}")
    print("-" * 50)
```

```python
#Soal 3
sql = """
SELECT 
    Product_line, 
    SUM(Kuantitas) AS jumlah_produk
FROM 
    productline.data_laprak_2
GROUP BY 
    Product_line
ORDER BY 
    jumlah_produk DESC;
"""
kursor1.execute(sql)
result_jumlah_produk = kursor1.fetchall()
print("\nJumlah produk terjual setiap kategori:")
print("-" * 50)
for row in result_jumlah_produk:
    print(f"Kategori: {row[0]}")
    print(f"Jumlah terjual: {row[1]}")
    print("-" * 50)
```

```python
#Soal 4
sql = """
SELECT 
    Product_Line, 
    Harga_Satuan,
    IF(Harga_Satuan BETWEEN 14.48 AND 42.79, 'Murah',
       IF(Harga_Satuan BETWEEN 42.79 AND 71.11, 'Cukup',
          IF(Harga_Satuan BETWEEN 71.11 AND 99.42, 'Mahal', 'Mahal Banget'))) AS Kategori_Harga
FROM 
    productline.data_laprak_2;
"""
kursor1.execute(sql)
result = kursor1.fetchall()
print("\nKategori Harga")
print("-" * 50)
for row in result:
    print(f"Kategori: {row[0]}")
    print(f"Harga Satuan: {row[1]}")
    print(f"Kategori Harga: {row[2]}")
    print("-" * 50)
```

```python
#Soal 5
sql = """
SELECT 
    Kota,
    CASE
        WHEN Kota = 'Yangon' THEN '001'
        WHEN Kota = 'Naypyitaw' THEN '002'
        WHEN Kota = 'Mandalay' THEN '003'
    END AS ID_Kota,
    SUM(Total) AS Jumlah_Total,
    CASE
        WHEN SUM(Total) BETWEEN 0 AND 6500 THEN 'Rendah'
        WHEN SUM(Total) BETWEEN 6500 AND 10000 THEN 'Sedang'
        ELSE 'Tinggi'
    END AS kategori_harga
FROM 
    productline.data_laprak_2
GROUP BY 
    Kota;
"""
kursor1.execute(sql)
result = kursor1.fetchall()
print("\nTotal Setiap Kota")
print("-" * 50)
for row in result:
    print(f"Kota: {row[0]}")
    print(f"ID Kota: {row[1]}")
    print(f"Jumlah Total: {row[2]}")
    print(f"Kategori Harga: {row[3]}")
    print("-" * 50)
```

```python
#Soal 6
sql = """
SELECT 
    Kota,
    SUM(CASE WHEN Kota = 'Yangon' THEN 1 ELSE 0 END) AS jumlah_transaksi_Yangon,
    SUM(CASE WHEN Kota = 'Naypyitaw' THEN 1 ELSE 0 END) AS jumlah_transaksi_Naypyitaw,
    SUM(CASE WHEN Kota = 'Mandalay' THEN 1 ELSE 0 END) AS jumlah_transaksi_Mandalay
FROM 
    productline.data_laprak_2
GROUP BY 
    Kota;
"""
kursor1.execute(sql)
result = kursor1.fetchall()
print("\nTotal transaksi setiap kota")
print("-" * 50)
print(f"Jumlah Transaksi Yangon: {int(result[0][1])}")
print(f"Jumlah Transaksi Naypyitaw: {int(result[1][2])}")
print(f"Jumlah Transaksi Mandalay: {int(result[2][3])}")
print("-" * 50)
```

```python
#Soal 7
sql = """
SELECT 
    Tipe_Anggota,
    AVG(Total) AS rata_rata_total_pembelian
FROM 
    productline.data_laprak_2
GROUP BY 
    Tipe_Anggota;
"""
kursor1.execute(sql)
result = kursor1.fetchall()
print("\nRata-rata pembelian setiap jenis anggota")
print("-" * 50)
for row in result:
    print(f"Tipe anggota: {row[0]}")
    print(f"Rata-rata: {row[1]:.2f}")
    print("-" * 50)
```

```python
#Soal 8
sql = """
SELECT 
    Kota,
    AVG(Total) AS rerata_total_pembelian
FROM 
    productline.data_laprak_2
WHERE 
    Payment = 'Ewallet' 
GROUP BY 
    Kota;
"""
kursor1.execute(sql)
result = kursor1.fetchall()
print("\nRata-rata total pembelian setiap kota menggunakan E-wallet")
print("-" * 50)
for row in result:
    print(f"Kota: {row[0]}")
    print(f"Rata-rata total pembelian: {row[1]:.2f}")
    print("-" * 50)
```
