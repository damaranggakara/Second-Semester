```python
# bubble sort
def bubble_sort(lst):
    n = len(lst)
    for i in range(n):
        #track pertukaran selama iterasi
        swapped = False
        #memeriksa elemen terakhir sebanyak i sudah di tempat yang benar
        for j in range(0, n - i - 1):
            #tukar bila elemen yang ditemukan lebih bsear dari elemen setelahnya
            if lst[j] > lst[j + 1]:
                lst[j], lst[j + 1] = lst[j + 1], lst[j]
                swapped = True
        #Jika tidak ada dua elemen yang ditukar oleh loop dalam, maka daftar sudah diurutkan
        if not swapped:
            break
bubble_sort(ruang)
# tampilkan hasil urutan
for value in ruang:
    print(value)
```

```python
#insertion sort
def insertion_sort(lst):
    for i in range(1, len(lst)):
        key = lst[i]
        j = i - 1
        #pindah lemen pertama yang lebih besar dari key ke posisi di depan
        while j >= 0 and key < lst[j]:
            lst[j + 1] = lst[j]
            j -= 1
        lst[j + 1] = key
insertion_sort(ruang)

#display sort
for value in ruang:
    print(value)
```

```python
#selection sort (cari angka ke terkecil pindah ke kiri)
def selection_sort(lst):
    n = len(lst)
    for i in range(n):
        # mencari elemen terkecil dari list yang belum terurut
        min_index = i
        for j in range(i + 1, n):
            if lst[j] < lst[min_index]:
                min_index = j
        # tuker elemen tersebut dengan elemen pertama
        lst[i], lst[min_index] = lst[min_index], lst[i]
selection_sort(ruang)

#display sort
for value in ruang:
    print(value)
```

```python
#merge sort (bagi dua)
def merge_sort(lst):
    if len(lst) <= 1:
        return lst
    #bagi list menjadi dua bagian
    mid = len(lst) // 2
    left_half = lst[:mid]
    right_half = lst[mid:]
    #sort dua bagian
    sorted_left = merge_sort(left_half)
    sorted_right = merge_sort(right_half)
    #gabung dua bagian
    return merge(sorted_left, sorted_right)

def merge(left, right):
    merged = []
    left_index = 1
    right_index = 1
    #gabung dua list dengan membandingkan elemen
    while left_index < len(left) and right_index < len(right):
        if left[left_index] <= right[right_index]:
            merged.append(left[left_index])
            left_index += 1
        else:
            merged.append(right[right_index])
            right_index += 1
    #tambahkan sia elemen ke kiri atau kanan jika ada
    merged.extend(left[left_index:])
    # Tambahkan sisa elemen dari right (jika ada)
    merged.extend(right[right_index:])
    return merged

merge_sort(ruang)

#display sort
for value in ruang:
    print(value)
```

```python
#quick sort 
#Proses utama dalam quick sort adalah partition(), target dari partisi adalah untuk menempatkan pivot (elemen apapun) pada posisi yang bener di sorted array dan pindahkan elemen lebih kecil ke sisi kiri dari pivot
#Pilih angka median ke paling kanan, trs tuker2 angka yang kecil dan besar, besar ke kanan dan kecil ke kiri sampai semuanya terurut. Lalu bagi dua kelompok antara kecil dan besar lalu tuker2 lagi. 
def quick_sort(lst):
    if len(lst) <= 1:
        return lst
    #pilih elemen pivot
    pivot = lst[len(lst) // 2]
    #bagi daftar menjadi tiga sublist
    left = [x for x in lst if x < pivot]
    middle = [x for x in lst if x == pivot]
    right = [x for x in lst if x > pivot]
    #urutkan secara rekursif sublist dan menggabungkannya dengan pivot
    return quick_sort(left) + middle + quick_sort(right)
quick_sort(ruang)

#display sort
for value in ruang:
    print(value)
```

```python
#heap sort
#cari elemen minimum dan pindahkan ia ke awal dan diulang terus menerus
#Ubah array menjadi tree (kiri ke kanan)
#Pindah angka terbesar ke atas tree menjadi root (max heap)
#Lalu pindah angka yang jadi root ke ujung tree agar ia pindah ke posisi terakhir array and then remove it
def heapify(lst, n, i):
    largest = i #Inisialisasi angka terbesar jadi root sebagai root
    left = 2 * i + 1  #anak kiri
    right = 2 * i + 2  #anak kanan
    #periksa anak mana yang lebih besar dari root
    if left < n and lst[left] > lst[largest]:
        largest = left
    #periksa anak kanan ada dan lebih besar dari terbesar saat ini
    if right < n and lst[right] > lst[largest]:
        largest = right
    #ganti root jika anak lebih besar
    if largest != i:
        lst[i], lst[largest] = lst[largest], lst[i]  #tukar
        #Secara rekursif membuat heap untuk sub-pohon yang terpengaruh
        heapify(lst, n, largest)

def heap_sort(lst):
    n = len(lst)
    #bangun maxheap.
    #mulai dari node daun terakhir dan membuat heap ke bawah hingga ke root
    for i in range(n // 2 - 1, -1, -1):
        heapify(lst, n, i)   
    #satu per satu tarik elemen dari heap dan mempertahankan sifat heap
    for i in range(n - 1, 0, -1):
        #tukar elemen terakhir dengan elemen pertama (elemen terbesar)
        lst[i], lst[0] = lst[0], lst[i]
        #perbaiki sifat maxheap pada sisa daftar (kecuali elemen terbesar yang telah dipindahkan)
        heapify(lst, i, 0)
heap_sort(ruang)

#display sort
for value in ruang:
    print(value) 
```
