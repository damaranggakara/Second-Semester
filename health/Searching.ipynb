```python
#imports
import pandas as pd
```

```python
# Original DataFrame
df = pd.read_csv("C:\\Users\\umay2\\Desktop\\SUFFER\\UNIV\\SEMESTER 2\\Struktur Data\\health\\DATA TUGAS SORT,SEARCH.csv")
print(df)
```

```python
#preprocessing
#Average Monthly Tax Credit (2016)
amtc = df['Average Monthly Tax Credit (2016)']
#hilangin $
amtc = amtc.replace('[\$]','', regex = True).astype(int)
print(amtc)
# Convert the column data to a list
amtc_list = amtc.tolist()
print(amtc_list)
```

```python
# Sequential search function
def sequential_search(data, target):
    for index, value in enumerate(data):
        if value == target:
            return index  # Return the index where the target is found
    return -1  # Return -1 if the target is not found

# Test the sequential search function
target_value = 7  # Replace with the value you are searching for
result_index = sequential_search(amtc_list, target_value)

if result_index != -1:
    print(f"Target value {target_value} found at index {result_index}.")
else:
    print(f"Target value {target_value} not found in the list.")

# Display the DataFrame and processed list
print(amtc_list)
```

```python
#Binary searching (iterative) (perbandingan median)
def binary_search_iterative(lst, target):
    left, right = 0, len(lst) - 1

    while left <= right:
        mid = left + (right - left) // 2  # Avoid potential overflow

        # Check if target is present at mid
        if lst[mid] == target:
            return mid
        # If target is greater, ignore the left half
        elif lst[mid] < target:
            left = mid + 1
        # If target is smaller, ignore the right half
        else:
            right = mid - 1

    # If we reach here, the element was not present
    return -1

# Example usage:
if __name__ == "__main__":
    my_list = amtc_list
    target_value = 7

    result = binary_search_iterative(my_list, target_value)

    if result != -1:
        print(f"Element {target_value} found at index {result}.")
    else:
        print(f"Element {target_value} not found in the list.")
```
