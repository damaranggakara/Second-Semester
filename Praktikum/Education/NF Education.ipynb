```python
#EDUCATION
#Import
import pandas as pd
import numpy as np
```

```python
# Read the Excel file into a pandas DataFrame
filedata = r"C:\\Users\\umay2\\Desktop\\SUFFER\\UNIV\\SEMESTER 2\\Basis Data\\PRAKTIKUM BASDAT\\Tugas 1\\Education\\student_excelvr_0nf.xlsx"
readexcel = pd.read_excel(filedata)
df= pd.DataFrame(readexcel)
print("DATA SEKOLAH:")
print(df)
```

```python
#1NF
#Cek bila ada data yang kosong
onenf= pd.isnull(df).sum()
print(onenf)
```

```python
#Isi kolum weekly study hours dengan 0
for i in range(len(df['Weekly_Study_Hours'])):
    if pd.isnull(df.loc[i, 'Weekly_Study_Hours'])==True:
        df.loc[i,'Weekly_Study_Hours']=0
#transfer hasil 1NF
df.to_excel("C:\\Users\\umay2\\Desktop\\SUFFER\\UNIV\\SEMESTER 2\\Basis Data\\PRAKTIKUM BASDAT\\Tugas 1\\Education\\student_excelvr_1nf.xlsx")
```

```python
#2NF
#Hilagkan partial dependency 
#membuat 2 kelompok kolum terdiri dari biodata sisawa dan laporan nilai
biodata = df[['Id','Student_Age','Sex',  'High_School_Type',  
              'Scholarship','Additional_Work','Sports_activity','Transportation','Report_Card_ID']]
laporan= df[['Report_Card_ID','Weekly_Study_Hours',  'Attendance','Reading','Notes'  
             ,'Listening_in_Class','Project_work','Grade']]
#Hasil 2NF
print("BIODATA SISWA:")
print(laporan)
print("LAPORAN SISWA:")
print(biodata)
#transfer hasil 2nf
with pd.ExcelWriter("C:\\Users\\umay2\\Desktop\\SUFFER\\UNIV\\SEMESTER 2\\Basis Data\\PRAKTIKUM BASDAT\\Tugas 1\\Education\\student_excelvr_2nf.xlsx") as writer:
    biodata.to_excel(writer, sheet_name="BIODATA SISWA")
    laporan.to_excel(writer, sheet_name="LAPORAN SISWA")
```

```python
#3NF
#Hilangkan transitive dependency
#cari relasi yang bisa di connect
idlaporangrade= df[['Id','Report_Card_ID','Grade']]
biodata = df[['Id','Student_Age','Sex',  'High_School_Type',  
              'Scholarship','Additional_Work','Sports_activity','Transportation']]
laporan= df[['Report_Card_ID','Weekly_Study_Hours',  'Attendance','Reading','Notes'  
             ,'Listening_in_Class','Project_work','Grade']]
#Hasil 3NF
print("HASIL SEMESTER SISWA:")
print(idlaporangrade)
print("BIODATA SISWA:")
print(laporan)
print("LAPORAN SISWA:")
print(biodata)
#transfer hasil 3NF
with pd.ExcelWriter("C:\\Users\\umay2\\Desktop\\SUFFER\\UNIV\\SEMESTER 2\\Basis Data\\PRAKTIKUM BASDAT\\Tugas 1\\Education\\student_excelvr_3nf.xlsx") as writer:
    idlaporangrade.to_excel(writer, sheet_name="NILAI SISWA")
    biodata.to_excel(writer, sheet_name="BIODATA SISWA")
    laporan.to_excel(writer, sheet_name="LAPORAN SISWA")
```
